```python
#Random Forest Regression~ version of ensemble learning
#steps~
#step1-Pick at random K data points from the Training set
#step2-Build the Decision Tree associated to these K data points
#step3-Choose the number Ntree of trees you want to build and repeat steps 1 and 2
#step4-For a new data point, make each one of your Ntree trees predict the value of Y to for the dat point in question, and 
#assign the new data point the average across all of the predicted Y values.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
#fitting random forest regression to the dataset
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X, y)
```

```python
y_pred = regressor.predict(6.5)
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid),1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title("Truth or bluff (Random forest regression)")
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
y_pred
```

```python
regressor1= RandomForestRegressor(n_estimators = 100,random_state = 0)
regressor1.fit(X, y)
y_pred = regressor1.predict(6.5)
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid),1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title("Truth or bluff (Random forest regression)")
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
#No. of intervals(steps) remains same but accuracy increases
```
