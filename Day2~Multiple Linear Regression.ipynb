```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
dataset = pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
#State is having catagorical data
#Dummy variables~New York and California
#We won't include both the dummy variables as D2=1-D1 ~ Dummy variable trap
#Always omit one dummy variable(We can have only one dummy variable in our model)
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
y = y.astype(int)
```

```python
y
```

```python
#for state column, we'll use onehotencoder
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X = LabelEncoder()
X[:, 3] = labelencoder_X.fit_transform(X[:, 3])
onehotencoder_X = OneHotEncoder(categorical_features = [3])
X = onehotencoder_X.fit_transform(X).toarray()
```

```python
X = X.astype(int)
X
```

```python
#Avoiding the dummy variable trap
X = X[:, 1:]
```

```python
X
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train
```

```python
X_test
```

```python
#feature scaling will be done by multiple linear regression library itself
#fitting multiple linear regression to the training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#predicting the test set results
y_pred = regressor.predict(X_test)
```

```python
#To have only those independent variables that have much more significance and to eliminate all those with negligieble significance, we'll use  backward elimination method
#Building the optimal model using Backward Elimination
import statsmodels.formula.api as sm
#we need to add one column x0 to the dataset for b0(constant) and value of x0 = 1
#X = np.append(arr = X, values = np.ones((50, 1)).astype(int), axis = 1) #axis  = 1 for adding column and axis = 0 for adding line
#it'll append all the ones in last
```

```python
X
```

```python
#to append all the ones at begining
X =  np.append(arr = np.ones((50,1)).astype(int), values = X, axis = 1)
```

```python
X
```

```python
X_opt = X[:,[0,1,2,3,4,5]]
#OLS~Ordinary Least Square
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
```

```python
#Steps for Backward Elimination-
#step1-selct a significance level to stay in the model(SL = 0.05)
#step2-fill the full model with all possible predictors
#step3-consider the predictor with the highest P-value. If P>SL, go to step4, otherwise model is ready
#step4-fit model without this variable
```

```python
#Implementing step3(P-value is probabaility)
regressor_OLS.summary()
```

```python
#As x2 is having highest p value which is greater than SL, we'll remove it and again fit the model with remaining variables
X_opt = X[:,[0,1,3,4,5]]
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
```

```python
X_opt = X[:,[0,3,4,5]]
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
```

```python
X_opt = X[:,[0,3,5]]
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
#As we can see adjusted R-squared increases, theresfore it's a better model
```

```python
X_opt = X[:,[0,3]]
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
#As we can see adjusted R-squared decreases, therefore it's not a better model
```

```python
#So, model with independent variables having index 0, 3 and 5 is best model
```
